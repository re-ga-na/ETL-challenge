```python
#converting csv of a dataset from kaggle of top spotify songs into pandas df
#https://www.kaggle.com/edumucelli/spotifys-worldwide-daily-song-ranking


import pandas as pd
from sqlalchemy import create_engine
```

```python
df = pd.read_csv("./SPOTIFY_DATA.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
#removing commas from track names
df["Track Name"] = df["Track Name"].map(lambda x: str(x).replace('"',"").replace("'","").replace(",",""))
```

```python
df.head()
```

```python
#spotify ids
spotifyid=[]
urllist=df["URL"]
#urllist
```

```python
for longurl in urllist:
    idnumber = longurl.rpartition('/')[-1]
    str(idnumber)
    spotifyid.append([idnumber])
    
#print(idnumber)
```

```python
df["Spotify ID"] = spotifyid

df.dtypes
```

```python
df.head()
```

```python
df.to_csv("./SPOTIFY_DATA.csv",index=False)
```

```python
df.rename(columns={"Position":"chart_position",
                    "Track Name":"track_name",
                    "Streams":"streams",
                    "URL":"url",
                    "Spotify ID":"spotifyid",
                    "Artist":"artist",
                    "Region":"region"},inplace=True)

df = df[['spotifyid', 'chart_position', 'track_name', 'streams', 'artist', 'region']].copy()
df.head()
```
